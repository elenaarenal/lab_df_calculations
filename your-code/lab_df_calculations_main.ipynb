```python
df=df.dropna() #quitar todas las filas que tienen al menos un valor nulo. CUIDADO
#chequear donde están los nulos, ver si tiene sentido quitarlos, etc.
df.head()
```

```python
#307645 original y 307477 filas después de eliminar las filas con valores nulos. Tiene sentido.
```

```python
#create the aggregates:

df.groupby('SUPPLIER').sum()

#no tiene sentido hacer la suma..¿18154 años??
```

```python
grupo_supplier = df.groupby('SUPPLIER').agg({'YEAR': 'max',
                                            'RETAIL SALES': 'sum',
                                             'RETAIL TRANSFERS':'sum', 
                                            'WAREHOUSE SALES': 'sum'
                                                })
grupo_supplier     

#así tiene más sentido: ver las ventas por supplier
```

```python
df.groupby('ITEM CODE').sum()
```

```python
grupo_item = df.groupby('ITEM CODE').agg({'YEAR': 'max',
                                            'RETAIL SALES': 'sum',
                                             'RETAIL TRANSFERS':'sum', 
                                            'WAREHOUSE SALES': 'sum',
                                              'ITEM TYPE': 'first'
                                                })
grupo_item  


```

```python

df.to_csv('clean.csv')
```

```python
grupo_item.to_csv('item.csv')
```

```python
grupo_supplier.to_csv("supplier.csv") 
```
